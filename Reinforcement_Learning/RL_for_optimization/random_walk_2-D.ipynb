```python
import random
import numpy as np
import pandas as pd

random.seed(1)
n_steps = 10000
x = np.zeros(n_steps)
y = np.zeros(n_steps)
x[0], y[0] = 0, 0

for t in range(n_steps-1):
    (dx, dy) = random.choices([(0,1), (0, -1), (1, 0), (-1, 0)])[0]
    x[t+1], y[t+1] = int(x[t]+dx), int(y[t]+dy)

df = pd.DataFrame({'x':x, 'y':y})
ax = df.plot.line(x='x', y='y', title='2D random walk', xlabel='x', ylabel='y', legend=False)
```
